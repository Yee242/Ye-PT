```python
import numpy as np
import matplotlib.pyplot as plt

# 检测系统中可用的中文字体并配置
try:
    # Windows 系统常用字体
    plt.rcParams["font.family"] = ["SimHei", "Microsoft YaHei", "FangSong"]
except:
    try:
        # macOS 系统常用字体
        plt.rcParams["font.family"] = ["PingFang SC", "Heiti TC", "STHeiti"]
    except:
        # Linux 系统常用字体
        plt.rcParams["font.family"] = ["WenQuanYi Micro Hei", "Heiti TC", "SimHei"]

plt.rcParams["axes.unicode_minus"] = False  # 正确显示负号

# 生成日期数据（保持原有逻辑）
year = np.arange(2000, 2021).astype(np.str_)
month = np.random.randint(1, 13, size=20).astype(np.str_)
day = np.random.randint(1, 31, size=20).astype(np.str_)
date = np.array([])

for i in range(20):
    a = np.array([year[i], month[i], day[i]])
    b = ['/'.join(a)]
    date = np.append(date, b)

sales = np.random.randint(500, 2000, size=len(date))

# 优化图表显示
plt.figure(figsize=(12, 6))
plt.plot(date, sales, color='#00A1FF', linewidth=2)
plt.title('销量随时间变化趋势', fontsize=14)
plt.xlabel('日期', fontsize=12)
plt.ylabel('销量', fontsize=12)

plt.xticks(
    range(0, len(date), 2),
    ['%s' % i for i in date][::2],
    rotation=45,
    ha='right',
    fontsize=10
)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.gca().spines[['top', 'right', 'left']].set_visible(False)
plt.tight_layout()
plt.show()
```
