```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

# 配置中文字体
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]  # 设置中文字体优先级
plt.rcParams["axes.unicode_minus"] = False  # 正确显示负号

# 配置部分 - 可根据实际情况修改
config = {
    'data_dir': r"D:\Ye-PT\day3\exercise_data",  # 数据文件夹路径
    'years': [2015, 2016, 2017],
    'encoding': 'utf-8',  # 文件编码，如遇到中文乱码可尝试 'gbk'
    'file_pattern': "{year}年国内主要城市年度数据.csv",  # 文件名模式
    'output_file': "北上广深经济数据.csv",  # 输出文件名
    'selected_cities': ["北京", "上海", "广州", "深圳"],  # 选定的城市
    
    # 列名映射 - 如果实际数据列名不同，可在此修改
    'column_mapping': {
        'city': '地区',  # 城市列名
        'year': '年份',  # 年份列名
        'gdp': '国内生产总值',  # GDP列名
        'hospital': '医院、卫生院数',  # 医院列名
        'retail': '社会商品零售总额'  # 零售总额列名
    }
}

def get_file_paths():
    """获取所有数据文件的路径"""
    file_paths = []
    for year in config['years']:
        file_name = config['file_pattern'].format(year=year)
        file_path = os.path.join(config['data_dir'], file_name)
        
        # 检查文件是否存在
        if os.path.exists(file_path):
            file_paths.append(file_path)
            print(f"找到文件: {file_path}")
        else:
            print(f"警告: 文件不存在 - {file_path}")
    
    if not file_paths:
        raise FileNotFoundError("没有找到任何数据文件")
    
    return file_paths

def read_and_merge_data(file_paths):
    """读取并合并所有数据文件"""
    dfs = []
    
    for file_path in file_paths:
        try:
            # 从文件名提取年份
            file_name = os.path.basename(file_path)
            year_match = [int(s) for s in file_name.split() if s.isdigit()]
            if not year_match:
                year = None
            else:
                year = year_match[0]
            
            # 读取数据
            df = pd.read_csv(file_path, encoding=config['encoding'])
            
            # 检查必要的列是否存在
            required_columns = list(config['column_mapping'].values())
            missing_columns = [col for col in required_columns if col not in df.columns]
            
            if missing_columns:
                print(f"警告: 文件 {file_name} 缺少以下列: {', '.join(missing_columns)}")
                continue
            
            # 如果年份不在数据中，从文件名添加
            if config['column_mapping']['year'] not in df.columns and year is not None:
                df[config['column_mapping']['year']] = year
            
            dfs.append(df)
            print(f"成功读取文件: {file_name}, 行数: {len(df)}")
            
        except Exception as e:
            print(f"读取文件 {file_path} 时出错: {e}")
            continue
    
    if not dfs:
        raise ValueError("无法读取任何数据文件")
    
    # 纵向合并数据
    merged_df = pd.concat(dfs, ignore_index=True)
    print(f"\n数据合并完成，总行数: {len(merged_df)}")
    
    # 显示数据基本信息
    print("\n数据基本信息:")
    merged_df.info()
    
    # 显示数据集行数和列数
    rows, columns = merged_df.shape
    
    # 检查是否有足够的数据进行分析
    if rows < 10 or columns < 5:
        print("\n警告: 数据量看起来异常小，可能存在问题")
        print("数据前几行预览:")
        print(merged_df.head().to_string())
    
    return merged_df

def process_data(merged_df):
    """处理数据并完成所有分析任务"""
    # 使用列名映射获取实际列名
    city_col = config['column_mapping']['city']
    year_col = config['column_mapping']['year']
    gdp_col = config['column_mapping']['gdp']
    hospital_col = config['column_mapping']['hospital']
    retail_col = config['column_mapping']['retail']
    
    # 2. 处理缺省值，填充为0
    merged_df.fillna(0, inplace=True)
    print("\n缺省值已填充为0")
    
    # 3. 按照年份聚合，求每年的国内生产总值
    annual_gdp = merged_df.groupby(year_col)[gdp_col].sum().reset_index()
    print("\n每年的国内生产总值:")
    print(annual_gdp)
    
    # 4. 计算每个城市2015-2017年GDP的年均增长率
    try:
        # 透视表，行是城市，列是年份，值是GDP
        pivot_df = merged_df.pivot(index=city_col, columns=year_col, values=gdp_col)
        
        # 计算年均增长率
        growth_rates = ((pivot_df[2017] / pivot_df[2015]) ** (1 / 2) - 1) * 100
        growth_rates = growth_rates.reset_index(name="年均增长率(%)")
        
        # 找出增长率最高和最低的五个城市
        top_five = growth_rates.sort_values("年均增长率(%)", ascending=False).head(5)
        bottom_five = growth_rates.sort_values("年均增长率(%)").head(5)
        
        print("\nGDP年均增长率最高的五个城市:")
        print(top_five)
        print("\nGDP年均增长率最低的五个城市:")
        print(bottom_five)
        
    except Exception as e:
        print(f"\n计算GDP年均增长率时出错: {e}")
        top_five = bottom_five = None
    
    # 5. 对医院、卫生院数进行归一化处理（Min-Max标准化）
    try:
        def min_max_normalize(df, column):
            """对指定列进行Min-Max标准化"""
            return (df[column] - df[column].min()) / (df[column].max() - df[column].min())
        
        # 按年份分组进行归一化
        merged_df[f"{hospital_col}_normalized"] = merged_df.groupby(year_col).apply(
            lambda x: min_max_normalize(x, hospital_col)).reset_index(0, drop=True)
        
        print(f"\n已对{hospital_col}进行归一化处理")
        
    except Exception as e:
        print(f"\n归一化处理时出错: {e}")
        merged_df[f"{hospital_col}_normalized"] = np.nan
    
    # 6. 提取北京、上海、广州、深圳四个城市2015-2017的GDP和社会商品零售总额数据
    try:
        selected_data = merged_df[
            merged_df[city_col].isin(config['selected_cities']) & 
            merged_df[year_col].isin(config['years'])
        ][[year_col, city_col, gdp_col, retail_col]]
        
        # 重命名列
        selected_data.columns = ["年份", "城市", "GDP", "社会商品零售总额"]
        
        # 保存到新的CSV文件
        output_path = os.path.join(config['data_dir'], config['output_file'])
        selected_data.to_csv(output_path, index=False, encoding="utf-8-sig")
        print(f"\n已将{', '.join(config['selected_cities'])}的数据保存到 {output_path}")
        
    except Exception as e:
        print(f"\n提取和保存选定城市数据时出错: {e}")
        selected_data = None
    
    return merged_df, annual_gdp, top_five, bottom_five, selected_data

def visualize_results(merged_df):
    """可视化分析结果"""
    # 使用列名映射获取实际列名
    city_col = config['column_mapping']['city']
    year_col = config['column_mapping']['year']
    gdp_col = config['column_mapping']['gdp']
    hospital_col = config['column_mapping']['hospital']
    
    try:
        # 1. GDP增长率可视化
        plt.figure(figsize=(14, 7))
        pivot_df = merged_df.pivot(index=city_col, columns=year_col, values=gdp_col)
        growth_rates = ((pivot_df[2017] / pivot_df[2015]) ** (1 / 2) - 1) * 100
        
        # 只显示增长率最高的15个城市
        top_15_cities = growth_rates.sort_values(ascending=False).head(15)
        ax = sns.barplot(x=top_15_cities.index, y=top_15_cities.values, palette='viridis')
        
        # 添加数据标签
        for p in ax.patches:
            height = p.get_height()
            ax.text(p.get_x() + p.get_width()/2., height + 0.5,
                    f'{height:.2f}%', ha='center')
        
        plt.title('GDP年均增长率最高的15个城市 (2015-2017)', fontsize=16)
        plt.xlabel('城市', fontsize=14)
        plt.ylabel('年均增长率 (%)', fontsize=14)
        plt.xticks(rotation=45, ha='right', fontsize=12)
        plt.tight_layout()
        
        # 保存图表
        chart_path = os.path.join(config['data_dir'], 'GDP_growth_rates.png')
        plt.savefig(chart_path, dpi=300, bbox_inches='tight')
        print(f"\n已保存GDP增长率图表到: {chart_path}")
        
    except Exception as e:
        print(f"\n可视化GDP增长率时出错: {e}")
    
    try:
        # 2. 医院资源变化可视化
        plt.figure(figsize=(14, 7))
        normalized_col = f"{hospital_col}_normalized"
        
        # 只显示部分主要城市
        major_cities = ["北京", "上海", "广州", "深圳", "重庆", "成都", "武汉", "杭州", "南京", "西安"]
        filtered_data = merged_df[merged_df[city_col].isin(major_cities)]
        
        ax = sns.lineplot(
            x=year_col, 
            y=normalized_col, 
            hue=city_col, 
            data=filtered_data,
            marker='o',
            markersize=8,
            err_style="bars",
            palette='deep'
        )
        
        plt.title('主要城市医院资源变化 (归一化后)', fontsize=16)
        plt.xlabel('年份', fontsize=14)
        plt.ylabel('归一化医院资源', fontsize=14)
        plt.legend(loc='upper right', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout()
        
        # 保存图表
        chart_path = os.path.join(config['data_dir'], 'hospital_resources.png')
        plt.savefig(chart_path, dpi=300, bbox_inches='tight')
        print(f"已保存医院资源变化图表到: {chart_path}")
        
    except Exception as e:
        print(f"\n可视化医院资源变化时出错: {e}")

def main():
    """主函数，执行整个数据分析流程"""
    print("=== 开始分析国内主要城市年度数据 ===")
    
    try:
        # 获取文件路径
        file_paths = get_file_paths()
        
        # 读取并合并数据
        merged_df = read_and_merge_data(file_paths)
        
        # 处理数据并完成分析
        results = process_data(merged_df)
        
        # 可视化结果
        visualize_results(merged_df)
        
        print("\n=== 分析完成 ===")
        
    except Exception as e:
        print(f"\n执行过程中发生严重错误: {e}")
        print("请检查数据文件格式和路径是否正确")

if __name__ == "__main__":
    main()
```
